```python
# Q4. In an array, Count Pairs with given sum

def find_pair_sum(nums,target):
    
    for i in range(len(nums) - 1):
        for j in range(i+1, len(nums)):
            if nums[i] + nums[j] == target:
                print('Pair Found: ',(nums[i],nums[j]))
                return
    
    print('Pair not Found')
            
if __name__ == '__main__':
    
    # nums = [8,7,2,5,3,1]
    nums = []
    n = int(input('Enter total numbers of array: '))
    for i in range(n):
        element = int(input('Enter element: '))
        nums.append(element)
    
    target = int(input('Enter value which need to be compared: '))
    
    print('Array elements are: ',nums)
    print('Sum of elements in array compared with number: ',target)
    
    find_pair_sum(nums,target)
```

```python
# Q5. Find duplicates in an array

def is_duplicate(array):
    if len(array) != len(set(array)):
        return True
    else:
        return False 
    
if __name__ == '__main__':
    
    # array = [5, 3, 5, 2, 1, 6, 6, 4]
    array = []
    n = int(input('Enter total numbers of array: '))
    for i in range(n):
        element = int(input('Enter element: '))
        array.append(element)
    
    print('Array elements are: ',array)
    
    duplicates = []
    non_duplicates = []
    
    for i in array:
        if i not in non_duplicates:
            non_duplicates.append(i)
        else:
            duplicates.append(i)
    
    if is_duplicate(array):
        print('Duplicates found in array: ',duplicates)
    else:
        print('No Duplicates found in array: ',non_duplicates)
```

```python
# Q6. Find the Kth largest and Kth smallest number in an array

def get_largest(array,k):
    array.sort()
    return array[-k]

def get_smallest(array,k):
    array.sort()
    return array[k-1]

if __name__ == '__main__':
    
    # l = [1,2,6,4,5,3]
    l = []
    n = int(input('Enter total numbers of array: '))
    for i in range(n):
        element = int(input('Enter element: '))
        l.append(element)
    
    print('Array elements are: ',l)
    
    k = int(input('Enter value of k : ')) # consider kth value as 3
    
    print(f'{k}th largest element is: ', get_largest(l, k))
    print(f'{k}th smallest element is: ', get_smallest(l, k))
```

```python
# Q8. Reverse a string using a stack data structure

def reverse(s):
    str = ''
    for i in s:
        str = i + str
    return str

if __name__ == '__main__':
    
    # string = 'Prep Bytes'
    string = input('Enter any string: ')
    
    print('Original String: ',string)
    
    print('Reversed string: ',reverse(string))
    
```

```python
# Q10. Implement a queue using the stack data structure

class Queue:

	def __init__(self):
		self.s1 = []
		self.s2 = []

	def enQueue(self, x):

		while len(self.s1) != 0:
			self.s2.append(self.s1[-1])
			self.s1.pop()

		self.s1.append(x)
  
		while len(self.s2) != 0:
			self.s1.append(self.s2[-1])
			self.s2.pop()

	def deQueue(self):

		if len(self.s1) == 0:
			return -1;

		x = self.s1[-1]
		self.s1.pop()
		return x

if __name__ == '__main__':

	q = Queue()
	q.enQueue(1)
	q.enQueue(2)
	q.enQueue(3)

	print(q.deQueue())
	print(q.deQueue())
	print(q.deQueue())
```
